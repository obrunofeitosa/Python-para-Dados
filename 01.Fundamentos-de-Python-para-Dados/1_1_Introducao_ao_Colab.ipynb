##1.1 - Introdução ao Colab

###O que é o Google Collab?

O Google Colaboratory (Colab) é uma ferramenta gratuita do Google para rodar código Python no navegador. Ideal para machine learning e análise de dados, oferece acesso a GPUs/TPUs sem instalação local. É um ambiente interativo para programar de forma simplificada.

### O que é um Notebook?

Um notebook, no contexto de ferramentas como o Google Colab (que se baseia na tecnologia Jupyter Notebook), é um ambiente interativo que combina código executável, texto explicativo formatado (Markdown), visualizações de dados e outros elementos multimídia em um único documento. Diferente de scripts tradicionais, os notebooks permitem executar o código em células individuais e visualizar os resultados imediatamente, facilitando a experimentação, a documentação do processo de análise ou desenvolvimento e o compartilhamento do trabalho de forma compreensível. É uma ferramenta essencial para cientistas de dados, pesquisadores e educadores.

### O que significa a extensão .ipynb?

A extensão `.ipynb` significa "IPython Notebook". IPython foi o projeto original que deu origem ao que hoje conhecemos como Jupyter Notebooks (e o Colab é baseado nessa tecnologia). Essencialmente, um arquivo com a extensão `.ipynb` é um arquivo de texto em formato JSON que armazena todo o conteúdo do seu notebook: o código das células, a saída da execução, o texto em Markdown, e metadados sobre o notebook. É o formato padrão para salvar e compartilhar notebooks interativos.
