##Aula 2 - Comentários

### Comentários em Python

Comentários são linhas no código que o interpretador Python ignora. Eles servem para adicionar notas, explicações ou desativar temporariamente partes do código sem removê-las. São essenciais para tornar seu código mais legível e compreensível, tanto para você no futuro quanto para outras pessoas.

Existem duas formas principais de adicionar comentários em Python:

1.  **Comentários de uma única linha:** Começam com o símbolo `#`. Tudo que vier depois do `#` naquela linha será considerado um comentário e ignorado pelo Python.

```python
    # Comentários de uma linha

    """
    Este é um comentário
    que abrange
    várias linhas.
    """
```
