## Aula 3 - Variáveis e Tipos de Dados

### O que são Dados e Variáveis?

Em programação, **dados** são as informações com as quais seu programa trabalha. Podem ser números, textos, valores verdadeiro/falso, etc.

Uma **variável** é como uma "caixa" ou um "rótulo" que você usa para armazenar um dado na memória do computador. Você dá um nome a essa caixa (o nome da variável) e pode colocar um valor (o dado) dentro dela. Depois, você pode usar o nome da variável para acessar ou modificar o dado armazenado.

É como se você tivesse uma caixa rotulada "idade" e colocasse o número 25 dentro dela. Sempre que você usar a palavra "idade" no seu código, o Python saberá que você está se referindo ao número 25.

Em Python, você não precisa declarar o tipo de dado que uma variável vai guardar; o Python descobre isso automaticamente. Para criar uma variável e atribuir um valor a ela, você usa o sinal de igual (`=`).

```python
# Exemplo de criação de variáveis
nome = "Maria"  # Variável 'nome' guarda o texto "Maria" (string)
idade = 30      # Variável 'idade' guarda o número inteiro 30 (int)
altura = 1.65   # Variável 'altura' guarda o número decimal 1.65 (float)
esta_estudando = True # Variável 'esta_estudando' guarda um valor verdadeiro (boolean)

# Você pode imprimir o valor das variáveis
print(nome)
print(idade)
print(altura)
print(esta_estudando)
```

### Verificando Tipos de Dados com `type()`

Em Python, você pode verificar o tipo de dado armazenado em uma variável usando a função built-in `type()`. Essa função retorna o tipo da variável ou do valor passado como argumento. É útil para entender com que tipo de informação você está trabalhando e para depurar seu código.

```python
# Exemplos usando a função type()
nome = "Maria"
idade = 30
altura = 1.65
esta_estudando = True

print(type(nome))
print(type(idade))
print(type(altura))
print(type(esta_estudando))
print(type("Olá, mundo!")) # Você também pode verificar o tipo de um valor literal
print(type(100))
```
